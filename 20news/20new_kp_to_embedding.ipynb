```python
from sentence_transformers import SentenceTransformer
import numpy as np
import torch
from torch import nn
```

```python
import json
with open("kp/20news_keyphrases_clean.json", "r") as f: #In vins server, it's at different location 
    lines = f.readlines()
pred_data = [json.loads(l) for l in lines]
```

```python
use_docs = 10000
pred_data = pred_data[-use_docs:]
```

```python
all_keyphrases = {}
total_keyphrases = 0
for i in pred_data:
    for j in i["pred"]:
        total_keyphrases+=1
        if j not in all_keyphrases:
            all_keyphrases[j] = 1
        else:
            all_keyphrases[j]= all_keyphrases[j]+1
```

```python
model_path = '../phrase-bert-model' #On vins server it's at different location
model = SentenceTransformer(model_path)
```

```python
keyphrase_embedding = {}
for keyphrase, freq in all_keyphrases.items():
    keyphrase_embedding[keyphrase] = model.encode([keyphrase]).tolist()
```

```python
import json
with open('1000_docs_keyphrase_embedding.json', 'w') as fp:
    json.dump(keyphrase_embedding, fp)
```
